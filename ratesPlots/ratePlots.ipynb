```python
import pandas as pd
import numpy as np
import pickle
import matplotlib.pyplot as plt
```

```python
# Import data
df = pd.read_csv('USDC.csv')
df.timestamp = pd.to_datetime(df.timestamp)
```

```python
# Plot
plt.figure(figsize=(16,9))
plt.plot(df.timestamp, df.variable_borrow_rate_comp)
plt.plot(df.timestamp, df.variable_borrow_rate_aave_v1)
plt.plot(df.timestamp, df.variable_borrow_rate_aave_v2)
plt.title('USDC')
plt.xlabel('Date')
plt.ylabel('Borrow APY')
plt.savefig('USDC.png', transparent=True)
```
